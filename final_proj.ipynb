```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
names =[
    'DATE', 'PRICE', 'DIVIDEND', 'EARNINGS', 'CPI', 'LIR', 
    'RP',  'RD', 'RE', 'PE10', 'MPC'
]

# read in our S&P data
df = pd.read_csv('data_csv.csv', header=0, names=names, na_values='?')
# remove na values
df = df.dropna()

# checking data
# df.head(6)
# num_samples = df.shape[0]
# num_attributes = df.shape[1]
# print(num_samples, num_attributes)
```

```python
# response vector of monthly % change
classifier = np.array(df['MPC'])
y = np.array(df['EARNINGS'])
X = np.array(df['RE'])
```

```python
col = np.where(classifier>=0,'g','r')
plt.scatter(X,y,c=col)
plt.xlabel('RE')
plt.ylabel('MPC')
plt.grid(True)
```
